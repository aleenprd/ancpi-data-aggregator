```python
import pandas as pd

# read the Excel file
df = pd.read_excel('data/statistica-aprilie-2017/cereri.xlsx')
display(df) 
```

```python
# drop the unnecessary columns/rows

df = pd.read_excel('data/statistica-aprilie-2017/cereri.xlsx', header=0, skiprows=2)
df = df.drop(df.columns[0], axis=1)
display(df)
```

```python
# rename the columns 

df.columns = ['county', 'request_type', 'online', 'window']
display(df)
```

```python
# fill in the missing values
df['county'] = df['county'].ffill()
display(df)
```

```python
# remove the rows that contain 'Total'

df = df[df['county'] != 'TOTAL']


display(df)
```

```python
# melt the dataframe to have a long format
df_melted = df.melt(id_vars=['county', 'request_type'], value_vars=['online', 'window'], var_name='request_method', value_name='number_of_requests')
display(df_melted)
```

```python
# unique request types values
unique_request_types = df_melted['request_type'].unique()
print(unique_request_types)
```

```python
# replace the request types with english names
df_melted['request_type'] = df_melted['request_type'].replace({
    'altele': 'other',
    'informare': 'information',
    'inscriere': 'registration',
    'receptie': 'reception',
})
display(df_melted.head())
```

```python
# save the dataframe to a CSV file
df_melted.to_csv('applications_table.csv', index=False)
```
