```python
import pandas as pd

#read the excel file
df = pd.read_excel('data/statistica-aprilie-2017/ipoteci.xlsx')
display(df)
```

```python
# drop index, total column, first row and total row

df = pd.read_excel(
    'data/statistica-aprilie-2017/ipoteci.xlsx',
    header=[1,2,3,4],
    skiprows=1,
    skipfooter=1
    )

df = df.drop(df.columns[[0, -1]], axis=1)

display(df)
```

```python
display(df.columns.to_list())
```

```python
# replace any entry that contains 'Unnamed' with an empty string

df.columns = pd.MultiIndex.from_tuples(
    tuple('' if 'Unnamed' in str(x) else x for x in tup)
    for tup in df.columns
)

display(df.head())
display(df.columns.to_list())
```

```python
# set variable english_names

english_columns = {
    'Județ': 'county',
    'Număr terenuri': 'plots',
    'Nr. Unități Individuale': 'individual_units',
    'Extravilan': 'rural',
    'Intravilan': 'urban',
    'Ipoteci Active': 'active_mortgages',
    'Ipoteci Radiate': 'cancelled_mortgages',
    'Agricole': 'agricultural',
    'Neagricole': 'non_agricultural',
    'Cu construcții': 'with_construction',
    'Fără construcții': 'without_construction'
}
```

```python
# apply the english names to the columns
df.columns = pd.MultiIndex.from_tuples(
    tuple(english_columns.get(str(x), x) for x in tup)
    for tup in df.columns
)
display(df.head())
display(df.columns.to_list())
```

```python
id_vars = [col for col in df.columns if col[0] == 'county']

df_melted = df.melt(id_vars=id_vars)
display(df_melted.head())
```

```python
df_melted.columns=['county', 'property_type', 'location_type', 'mortgage_status', 'usage_type', 'quatity']
display(df_melted.head())
```

```python
# insert month and year columns
df_melted.insert(0, 'month_number', 4)
df_melted.insert(1, 'month', 'april')
df_melted.insert(2, 'year', 2017)

display(df_melted)
```

```python
# save the dataframe to a csv file
df_melted.to_csv('mortgage_table.csv', index=False)
```
